Inspecting the data for its info, datatypes, null values, description, duplicates

```
#to inspect the data
df.info()
```

```
#replacing the blanks with 0 as tenure is 0 and change datatype for TotalCharges to float from object
df["TotalCharges"] = df["TotalCharges"].replace(["  "," "],"0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```
#check for duplicates for unique value
df["customerID"].duplicated().sum()
```

```
#converting 0 and 1 values in SeniorCitizen column to "yes" or "no" for better understanding
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```
df.head()
```

Analysing data (why curn out of customer, types of customers etc.)

```
ax = sns.countplot(x = df['Churn'], data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show
```

```
gb = df.groupby("Churn").agg({'Churn':'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("% of Churned Customers")
plt.show()
```

From the given pie-chart we can conclude that 26.54% customers have churned out. Now let's explore the reason behind it.

```
#on the basis of gender
ax = sns.countplot(x="gender", data = df, hue="Churn")
plt.title("Churn Distribution by Gender")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

Since the number of male and female customers churning out is equal, the reason for churning out of customers is not based on gender, hence requires further investigation.

```
#visualising how many customers are senior citizens
ax = sns.countplot(x="SeniorCitizen", data = df)
plt.title("Churn Distribution by Age Limit (if senior citizen or not)")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```
# 1. Crosstab: counts and row-wise percentages
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"])          # raw counts
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100               # % within each SeniorCitizen group

# 2. Stacked bar plot
fig, ax = plt.subplots(figsize=(6, 4))

x = np.arange(len(ct_pct.index))
bottom = np.zeros(len(ct_pct))

for col in ct_pct.columns:
    heights = ct_pct[col].values
    ax.bar(x, heights, bottom=bottom, label=col)
    bottom += heights

# 3. Add % labels in the middle of each segment
for i, sc in enumerate(ct_pct.index):
    cum_height = 0
    for col in ct_pct.columns:
        h = ct_pct.loc[sc, col]
        if h > 0:
            ax.text(i, cum_height + h / 2, f"{h:.1f}%",
                    ha="center", va="center", fontsize=9)
        cum_height += h

ax.set_xticks(x)
ax.set_xticklabels(ct_pct.index)
ax.set_ylabel("Percentage of customers")
ax.set_title("Churn Distribution by Age Limit (Senior citizen or not)")
ax.set_ylim(0, 120)
ax.legend(title="Churn")
plt.show()
```

Observation: Comparatively a greater percentage of people in senior citizen category have churned out.
We'll now check churned out on the basis of tenure.

```
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins = 72, hue = "Churn")
plt.show()
```

Observations: Churn out by customers in the initial days, especially within 1st month. Therefore, people who have used our services for a long time have stayed and people who have used our services have churned out. This could be due to the kind of contract people have opted for. Hence, analysing the churn out distribution w.r.t the type of contract opted by customer.  

```
#visualising the type of contract opted for by the customer
ax = sns.countplot(x="Contract", data = df, hue = "Churn")
plt.title("Count of Customers by Contract")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

Observation: Customers on month-to-month contract are likely to churning out more then from those who have 1-year or 2-year of contract.
Solution: Showcasing the benefits of 1-year and 2-year contract to the customers.

```
import math

cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

n_cols = 3                                   # subplots per row
n_plots = len(cols)
n_rows = math.ceil(n_plots / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()  # make it 1D for easy indexing

for i, col in enumerate(cols):
    ax = axes[i]
    sns.countplot(data=df, x=col, hue="Churn", ax=ax)
    ax.set_title(col)
    ax.set_xlabel("")          # optional: cleaner look
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=20)

# Hide any unused subplots (if n_plots not multiple of n_cols)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# One shared legend (from the first axis)
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title="Churn", loc="upper right")

plt.tight_layout()
plt.show()
```

Observations: Most customers have phone and internet services, and these groups also contain the vast majority of churners.
Fiber optic internet users, and those without OnlineSecurity, OnlineBackup, DeviceProtection, or TechSupport, show noticeably higher churn than their counterparts.
Customers with â€œNo internet serviceâ€ or â€œNo phone serviceâ€ are fewer overall and have relatively low churn counts.
StreamingTV and StreamingMovies subscribers appear slightly more prone to churn than customers who have internet but donâ€™t use these streaming add-ons.
